```python

# AsegÃºrate de que sale_amount es numÃ©rico
dfventas['sale_amount'] = pd.to_numeric(dfventas['sale_amount'], errors='coerce')

# AsegÃºrate de que data_recorded es tipo datetime
dfventas['data_recorded'] = pd.to_datetime(dfventas['data_recorded'], errors='coerce')

# Extraer el aÃ±o
dfventas['Year'] = dfventas['data_recorded'].dt.year

# Agrupar por aÃ±o y sumar correctamente
ingreso_anual = dfventas.groupby('Year')['sale_amount'].sum().reset_index(name='ingreso_total')

# Mostrar resultados
print(ingreso_anual)
```

```python
import matplotlib.pyplot as plt

# Crear figura y eje
fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

# Graficar barras horizontales
ax.barh(ingreso_anual['Year'], ingreso_anual['ingreso_total'], color='skyblue')

# ConfiguraciÃ³n del grÃ¡fico
ax.set_xlabel("Ingreso Total ($)", fontsize=12, fontweight="bold")
ax.set_ylabel("AÃ±o", fontsize=12, fontweight="bold")
ax.set_title("GrÃ¡fico 1: Ingreso Anual por Venta de Propiedades", fontsize=14, fontweight="bold")

# Ajustar eje X para que las barras no queden pegadas al borde
ax.set_xlim(0, ingreso_anual['ingreso_total'].max() * 1.1)

# Agregar etiquetas de valor al final de cada barra
for i, valor in enumerate(ingreso_anual['ingreso_total']):
    ax.text(valor + (ingreso_anual['ingreso_total'].max() * 0.01), ingreso_anual['Year'][i], 
            f"${valor:,.2f}", va='center', fontsize=9)

# Ajustar mÃ¡rgenes
plt.tight_layout()

# Guardar el grÃ¡fico
fig.savefig("Grafico_Ingreso_Anual_Ventas_Horizontal", dpi=300, bbox_inches='tight')

# Mostrar el grÃ¡fico
plt.show()
```

```python
def agrupaYcuenta(df, columna, ordenar):
    return df.groupby(columna).size().reset_index(name='cantidad').sort_values(by='cantidad', ascending=ordenar)
```

```python
def agrupaYcuenta(dfParam, nomColumna, orden):
    gf = dfParam.groupby([nomColumna]).count()
    gf = gf.iloc[:, 0:1]  # Selecciona solo la primera columna
    gf.columns = ['cantidad']  # Renombra esa columna sin importar su nombre original
    gf = gf.sort_values("cantidad", ascending=orden)
    return gf
```
